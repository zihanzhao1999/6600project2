```python
import pandas as pd
import numpy as np
```

```python
#Read file
file_to_load = "incident_event_log.csv"
df = pd.read_csv(file_to_load)
```

```python
#Add the duration of the time used solving incident
df['openedtime']= pd.to_datetime(df['opened_at'])
df['closedtime']= pd.to_datetime(df['closed_at'])
df['time_period'] = df['closedtime'] - df['openedtime']
```

```python
#replace all ? value into NA
df = df.replace({'?': np.nan})
```

```python
#Find the missing value of each column
df.isnull().sum()
```

```python
#data clean
df.drop(columns = ['cmdb_ci','problem_id','rfc','vendor','caused_by'],inplace = True)
```

```python
#Set new column based on impact and urgency
#The new column Severity is aim to campare with the old column priority.
df["Severity"] = "SpecialIncident"
df["Severity"][(df["impact"] == "1 - High") & (df["urgency"] == "1 - High")] = "Critical"
df["Severity"][(df["impact"] == "2 - Medium") & (df["urgency"] == "1 - High")] = "High" 
df["Severity"][(df["impact"] == "1 - High") & (df["urgency"] == "2 - Medium")] = "High" 
df["Severity"][(df["impact"] == "2 - Medium") & (df["urgency"] == "2 - Medium")] = "Moderate" 
df["Severity"][(df["impact"] == "3 - Low") & (df["urgency"] == "2 - Medium")] = "Low" 
df["Severity"][(df["impact"] == "2 - Medium") & (df["urgency"] == "3 - Low")] = "Low"
df["Severity"][(df["impact"] == "3 - Low") & (df["urgency"] == "3 - Low")] = "Low"
```

```python
df.to_csv ('incident.csv', index = False, header=True)
```
